```python
# 전처리된 데이터를 가져온다.
import pandas as pd
finaljoined = pd.read_csv('finaljoined.csv',index_col = 0)
```

```python
# user에 대한 정보만 남기고, DataFrame을 정리한다.
userlist = finaljoined.sort_values(
    by='insert_time',ascending = False
).drop(
    [
        'application_id','loanapply_insert_time',
        'bank_id','product_id','loan_limit',
        'loan_rate','is_applied','insert_time'
    ],
    axis = 1
).drop_duplicates(subset = ['user_id']).reset_index(drop = True)
```

```python
# user에 대한 정보를 userlist.csv로 내보낸다.
userlist.to_csv('userlist.csv')
```
