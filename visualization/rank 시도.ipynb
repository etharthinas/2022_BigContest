```python
see4 = final[['application_id','user_id','is_applied','gender','houseown_type','existing_loan_cnt','income_type','employment_type','purpose','credit_score']]
see4 = see4.sort_values(by=['user_id'],ascending=True)
```

```python
see5 = see4.groupby('application_id').sum()
see5['apid'] = list(see5.index)
see5 = see5.reset_index(drop=True)
see5.rename(columns = {'apid':'application_id'}, inplace = True)
see5['applied']=see5['is_applied'].apply(lambda x: int(bool(x)))
see6 = pd.merge(see4[['user_id','application_id']].drop_duplicates(),see5[['application_id','applied']],on='application_id')
```
